```python
#creating range of values in array
import numpy as np
np.arange(1,10,2)
np.linspace(0,1,5)
```

```python
#indexing and slicing for 1d array
import numpy as np
arr=np.array([10,20,30,40])
print(arr[0])
print(arr[1:3])
print(arr[-1])
```

```python
#indexing and slicing for 2D array
import numpy as np
arr2=np.array([[10,20,30],[40,50,60]])
print(arr2[0,0])
print(arr2[0,2])
print(arr2[1,2])
```
